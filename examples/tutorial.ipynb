# yungdiagram tutorial

This notebook introduces the `yungdiagram` package — a Python library for working with Young diagrams, Young tableaux, and skew shapes.

These combinatorial objects appear throughout algebraic combinatorics and representation theory: in the representation theory of the symmetric group $S_n$ and general linear group $GL_n$, in the study of symmetric functions, and in a wide range of bijective and enumerative results.

**References:**

- Stanley, _Enumerative Combinatorics_, Vol. 2, Chapter 7
- Sagan, _The Symmetric Group_, Chapter 4
- Fulton, _Young Tableaux_


```python
from yungdiagram import Cell, YoungDiagram, SkewDiagram, YoungTableau, SkewTableau
```

## 1. Partitions and Young diagrams

An **integer partition** of $n$ is a weakly decreasing sequence of positive integers $\lambda = (\lambda_1 \geq \lambda_2 \geq \cdots \geq \lambda_k > 0)$ with $\sum \lambda_i = n$. We write $\lambda \vdash n$.

The **Young diagram** of $\lambda$ is the collection of boxes arranged in left-justified rows, with $\lambda_i$ boxes in row $i$.

There are three standard drawing conventions:

- **English**: row 1 at the top (most common in combinatorics)
- **French**: row 1 at the bottom (common in geometry)
- **Russian**: diagram rotated 45°, so rows become diagonals


```python
lam = YoungDiagram([4, 3, 1])
print(f"partition: {lam.partition}")
print(f"size (n):  {lam.size}")
```

```python
# English convention (default)
print("English:")
print(lam)
print()
print("French:")
print(f"{lam:french}")
print()
print("Russian:")
print(f"{lam:russian}")
```

```python
# In Jupyter, _repr_html_ renders an HTML table automatically
lam
```

## 2. Basic properties

### Conjugate partition

The **conjugate** (or transpose) $\lambda'$ of $\lambda$ is obtained by reflecting the diagram across its main diagonal: $\lambda'_j = |\{i : \lambda_i \geq j\}|$.


```python
lam_conj = lam.conjugate()
print(f"conjugate partition: {lam_conj.partition}")
print(lam_conj)
```

```python
# Self-conjugate partitions correspond to partitions into distinct odd parts
sc = YoungDiagram([3, 2, 1])
print(f"(3,2,1) is self-conjugate: {sc.is_self_conjugate()}")

sc2 = YoungDiagram([3, 1, 1])
print(f"(3,1,1) is self-conjugate: {sc2.is_self_conjugate()}")
```

### Durfee square

The **Durfee square** of $\lambda$ is the largest $d \times d$ square that fits in the top-left corner of the diagram. Its side length equals the number of parts $\lambda_i \geq i$.


```python
print(f"Durfee square size: {lam.durfee_square_size}")
print(lam.durfee_square())
```

```python
# Strict partitions: all parts are distinct
print(YoungDiagram([5, 3, 1]).is_strict())  # True
print(YoungDiagram([4, 3, 3]).is_strict())  # False
```

## 3. Young's lattice

Partitions are partially ordered by **containment**: $\mu \subseteq \lambda$ if $\mu_i \leq \lambda_i$ for all $i$. The covering relation is: $\mu \lessdot \lambda$ iff $\lambda$ is obtained from $\mu$ by adding exactly one box in a valid position (i.e., the result is still a partition). This gives **Young's lattice**.

A separate partial order is the **dominance order**: $\mu \preceq \lambda$ if $\sum_{i=1}^k \mu_i \leq \sum_{i=1}^k \lambda_i$ for all $k$.


```python
d = YoungDiagram([3, 2])

print("Addable cells (positions where a box can be added):")
print(d.addable_cells())
print()
print(d.draw_addable())
```

```python
print("Removable cells (corner boxes that can be deleted):")
print(d.removable_cells())
print()
print(d.draw_removable())
```

```python
# All diagrams reachable by adding one box
d.reachable_young_diagrams_by_addition()
```

```python
# Containment and dominance
mu = YoungDiagram([2, 1])
print(f"(3,2) contains (2,1): {d.contains(mu)}")
print(f"(3,2) dominates (2,2,1): {d.dominates(YoungDiagram([2, 2, 1]))}")
```

## 4. Hook lengths and counting standard Young tableaux

For a cell $(x, y)$ in a Young diagram (0-indexed, $x$ = column, $y$ = row):

- **arm length** $a(c)$: number of cells strictly to the right in the same row
- **leg length** $\ell(c)$: number of cells strictly below in the same column
- **hook length** $h(c) = a(c) + \ell(c) + 1$

The **hook-length formula** (Frame–Robinson–Thrall, 1954) counts standard Young tableaux:
$$f^\lambda = \frac{n!}{\prod_{c \in \lambda} h(c)}$$


```python
lam = YoungDiagram([4, 3, 1])
corner = Cell(0, 0)  # top-left cell

print(f"arm length of (0,0):  {lam.arm_length(corner)}")
print(f"leg length of (0,0):  {lam.leg_length(corner)}")
print(f"hook length of (0,0): {lam.hook_length(corner)}")
```

```python
# Number of SYTs for various shapes
examples = [
    [4],  # one row: always 1
    [1, 1, 1, 1],  # one column: always 1
    [2, 2],  # 2x2 square: 2
    [3, 2],  # staircase: 5
    [4, 3, 1],  # (4,3,1): 70
    [3, 2, 1],  # staircase: 16
]

for p in examples:
    d = YoungDiagram(p)
    print(f"f^{d.partition} = {d.number_of_standard_tableaux()}")
```

```python
# The diagonal content of a cell is x - y
# Cells on the main diagonal have content 0
for y, row in enumerate(lam.cells):
    for cell in row:
        print(f"cell {cell}: content {cell.content}")
```

## 5. Young tableaux

A **filling** of a Young diagram $\lambda$ assigns a positive integer to each box. A filling is:

- **semistandard** (SSYT): weakly increasing along rows left-to-right, strictly increasing down columns
- **standard** (SYT): uses each of $\{1, \ldots, n\}$ exactly once (automatically strictly increasing in both rows and columns)

The **weight** (or **content**) of a semistandard tableau is the tuple $(a_1, a_2, \ldots)$ where $a_i$ is the number of times $i$ appears.


```python
# A semistandard Young tableau of shape (4,3,1) and weight (2,2,2,1,1)
ssyt = YoungTableau([[1, 1, 2, 3], [2, 3, 4], [5]])
print(f"is_semistandard: {ssyt.is_semistandard()}")
print(f"is_standard:     {ssyt.is_standard()}")
print(f"weight:          {ssyt.weight}")
print()
print(ssyt)
```

```python
# In Jupyter, rendered as an HTML table
ssyt
```

```python
# A standard Young tableau
syt = YoungTableau([[1, 2, 5, 7], [3, 4, 6], [8]])
print(f"is_standard: {syt.is_standard()}")
print(syt)
```

```python
# French convention
print(f"{syt:french}")
```

```python
# Conjugate of a tableau: transpose the filling
# Conjugate of an SYT is an SYT of the conjugate shape
conj = syt.conjugate()
print(conj)
print(f"shape:\n{conj.diagram}")
print(f"is_standard: {conj.is_standard()}")
```

```python
# Tableaux support equality and hashing
t1 = YoungTableau([[1, 2], [3]])
t2 = YoungTableau([[1, 2], [3]])
t3 = YoungTableau([[1, 3], [2]])
print(t1 == t2)  # True
print(t1 == t3)  # False
print(len({t1, t2, t3}))  # 2
```

## 6. Skew diagrams and tableaux

Given partitions $\mu \subseteq \lambda$, the **skew shape** $\lambda/\mu$ is the set of boxes in $\lambda$ but not in $\mu$.

A skew shape is a:

- **horizontal strip** if it has at most one box in each column (equivalently, $\mu_i \geq \lambda_{i+1}$ for all $i$)
- **vertical strip** if it has at most one box in each row (equivalently, $\lambda_i - \mu_i \leq 1$ for all $i$)

These appear in the **Pieri rules** for multiplying Schur functions: $s_\lambda \cdot h_k = \sum_{\mu} s_\mu$ summed over $\mu/\lambda$ a horizontal strip of size $k$.


```python
big = YoungDiagram([4, 3, 2])
small = YoungDiagram([2, 1])
skew = SkewDiagram(big, small)

print(f"size: {skew.size}")
print()
print(skew)
```

```python
skew
```

```python
# Check strip conditions
hs = SkewDiagram(YoungDiagram([4, 3, 2]), YoungDiagram([3, 2]))
vs = SkewDiagram(YoungDiagram([3, 2, 1]), YoungDiagram([2, 2]))

print(f"horizontal strip: {hs.is_horizontal_strip()}")  # True
print(f"vertical strip:   {vs.is_vertical_strip()}")  # True
print(f"connected:        {skew.is_connected()}")
```

```python
# Skew tableaux: None marks the inner shape cells
st = SkewTableau([[None, None, 1, 2], [None, 2, 3], [1, 3]])
print(f"is_semistandard: {st.is_semistandard()}")
print(f"weight:          {st.weight}")
print()
print(st)
```

```python
st
```

```python
# Conjugate of a skew shape and tableau
print(skew.conjugate())
print()

sst = SkewTableau([[None, 1, 2], [None, 3, 4]])
print(f"is_standard: {sst.is_standard()}")
print(f"conjugate is standard: {sst.conjugate().is_standard()}")
```

## 7. Random Young diagrams

`YoungDiagram.random(n)` returns a diagram drawn uniformly at random from all partitions of $n$, using a dynamic programming approach on partition counts.


```python
d = YoungDiagram.random(20)
print(d)
```

## What's coming

The following features are planned for upcoming releases:

- **RSK correspondence** — Robinson-Schensted-Knuth row insertion: a word $w$ maps to a pair of tableaux $(P, Q)$ where $P$ is semistandard (insertion tableau) and $Q$ is standard (recording tableau). This is arguably the most important algorithm in the field.
- **Jeu de taquin** — forward and backward slides on skew tableaux; rectification (applying slides until no inner corners remain). Central to proofs of the Littlewood-Richardson rule.
- **Promotion and evacuation** — Schützenberger's operators on SYTs; fundamental to cyclic sieving phenomena and homomesy.
- **Tableau statistics** — descent set, major index, charge and cocharge (Lascoux-Schützenberger).
- **Enumeration** — list all SYTs of a given shape (rather than just counting them).
